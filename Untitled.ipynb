```python
%pylab inline
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import statsmodels.tsa.stattools as ts
import statsmodels.formula.api as sm
from pandas.stats.api import ols
```

```python
BTCprice = pd.read_csv('~/Dropbox/chives/data/bitfinex/candle/bitfinex_candle_tBTCUSD_1m.csv', index_col=0)
ETHprice = pd.read_csv('~/Dropbox/chives/data/bitfinex/candle/bitfinex_candle_tETHUSD_1m.csv', index_col=0)
BTCfund = pd.read_csv('~/Dropbox/chives/data/bitfinex/funding/bitfinex_funding_fBTC_1m.csv', index_col=0)
ETHfund = pd.read_csv('~/Dropbox/chives/data/bitfinex/funding/bitfinex_funding_fETH_1m.csv', index_col=0)
```

```python
ETHprice.index = pd.to_datetime(ETHprice.index, unit='ms')
ETHprice.index
```

```python
fig,ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.set_xlim([1485e9, 1493.5e9])
ax1.plot(ETHprice['close'], 'b-')
ax2.plot(ETHprice['volume'], 'r-')

ax1.set_ylabel('price')
ax2.set_ylabel('Volume')
plt.show()
```

```python
olsmodel = sm.ols(formula= 'ETHprice.close ~ ETHprice.volume + ETHfund.close + ETHfund.volume').fit()
print(olsmodel.summary())
```

```python
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.stattools import acf, pacf
open_log = np.log(ETHprice['close'])
open_log_diff = open_log - open_log.shift()
open_log.plot()
lag_acf = acf(open_log, nlags=20)
lag_pacf = pacf(open_log, nlags=20, method='ols')
```

```python
#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(open_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(open_log_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
```

```python
#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(open_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(open_log_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```
